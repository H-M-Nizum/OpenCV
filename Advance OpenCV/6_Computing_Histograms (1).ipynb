# Computing Histograms

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load image
img = cv.imread('../Resources/Photos/cats.jpg')

# Display image
cv.imshow('IMG', img)

# wait and destroy
cv.waitKey(0)
cv.destroyAllWindows()
```

## Histograms
    Histoframs basically allow to visualize the distribution of pixel intersities in and image.
	So whether it's a color images, or whether it's grayscale image, i can visualize these pixel intensity
	distribution with the help of a histogram, which is kind of like a graph or a plot that 
	will give a high level intuition of the pixel distribution in the image.

### 1. Histograms for grayscale images
    calcHist(images[it's list], channels, mask, histSize, ranges)
	channels -> Specify the index [0]
	mask -> None
	histSize -> The number of bins that we want to use for computing the histogram.[256]
	ranges -> [0, 256]

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('GRAYSCALE IMAGE', gray)

# Grayscale Histograms
gray_hist = cv.calcHist([gray], [0], None, [256], [0, 256])

plt.figure()
plt.title('Histogram for GrayScale Image')
plt.xlabel("Bins")
plt.ylabel("Number of pixels")
plt.plot(gray_hist)
plt.xlim(0, 256)
plt.show()

# wait and destroy
cv.waitKey(0)
cv.destroyAllWindows()
```
